```python
# natural language processing: n-gram ranking
import re
import unicodedata
import nltk
from nltk.corpus import stopwords
# add appropriate words that will be ignored in the analysis
ADDITIONAL_STOPWORDS = ['gender', 'equality', 'sexual', 'orientation', 'race', 'religion', 'sex', 'national', 'origin', 'opportunity', 'regard', 'color', 'veteran', 'protected', 'status']

import matplotlib.pyplot as plt
```

```python
def basic_clean(text):
  """
  A simple function to clean up the data. All the words that
  are not designated as a stop word is then lemmatized after
  encoding and basic regex parsing are performed.
  """
  wnl = nltk.stem.WordNetLemmatizer()
  stopwords = nltk.corpus.stopwords.words('english') + ADDITIONAL_STOPWORDS
  text = (unicodedata.normalize('NFKD', text)
    .encode('ascii', 'ignore')
    .decode('utf-8', 'ignore')
    .lower())
  words = re.sub(r'[^\w\s]', '', text).split()
  return [wnl.lemmatize(word) for word in words if word not in stopwords]
```

```python
jdFull_words = basic_clean(''.join(str(data['jdFull'].tolist())))
```

```python
jdBullet_words = basic_clean(''.join(str(data['jdBullet'].tolist())))
```

```python
(pd.Series(nltk.ngrams(jdFull_words, 1)).value_counts())[:30]
```

```python
unigram_data = (pd.DataFrame(nltk.ngrams(jdBullet_words, 1)).value_counts())[:50]
unigram_data.to_csv('/content/drive/MyDrive/Colab Notebooks/unigram_data.csv')
```

```python
bigram_data = (pd.DataFrame(nltk.ngrams(jdBullet_words, 2)).value_counts())[:50]
bigram_data.to_csv('/content/drive/MyDrive/Colab Notebooks/bigram_data.csv')
```

```python
trigram_data = (pd.DataFrame(nltk.ngrams(jdBullet_words, 3)).value_counts())[:50]
trigram_data.to_csv('/content/drive/MyDrive/Colab Notebooks/trigram_data.csv')
```
